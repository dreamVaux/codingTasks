```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv')
```

### 1. Boxplot

```python
# Boxplot for revs per mile
selected_data = car_data[car_data['Manufacturer'].isin(['Audi', 'Hyundai', 'Suzuki', 'Toyota'])]
sns.boxplot(x = 'Manufacturer', y = 'Rev.per.mile', data = selected_data)
plt.title('Boxplot of Revs per Mile for Selected Manufacturers')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Toyota

### 2. Histogram

```python
# Histogram of MPG
sns.histplot(car_data['MPG.city'], color = 'red', label = 'City MPG')
sns.histplot(car_data['MPG.highway'], color = 'blue', label = 'Highway MPG')
plt.title('Histogram of MPG: City v.s. Highway')
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
Highway

By examining the histograms representing miles per gallon (MPG) in urban and highway settings, we note that the distribution of MPG on highways typically surpasses that in cities. This implies that vehicles tend to operate more efficiently in terms of fuel consumption on highways than in urban driving environments.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(x = 'Wheelbase', y = 'Turn.circle', data = car_data)
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.title('Relationship between Wheelbase and Turning Circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  

The relationship is positive correlation. As wheelbase increases, turning circle increases.


### 4. Barplot

```python
# Barplot for mean horsepower
horsepower_mean = car_data.groupby('Type')['Horsepower'].mean().reset_index()
sns.barplot(x = 'Type', y = 'Horsepower', data = horsepower_mean)
plt.title('Mean Horsepower by Car Type')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
Yes, the barplot shows that larger cars have more horsepower.

The bar chart illustrating the average horsepower for various car categories (such as small, midsize, large, etc.) indicates that larger car types, like large vehicles, generally exhibit greater average horsepower in contrast to smaller ones. This indicates a positive correlation between car size and horsepower within this dataset.
